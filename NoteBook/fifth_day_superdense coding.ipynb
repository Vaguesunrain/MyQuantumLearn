# Superdense Coding

Generally,we can use a qubit  to replace tow classical bits at the cost of an e-bit.

```python
from qiskit import QuantumCircuit, QuantumRegister, Aer, transpile, assemble, execute

c = '1'
d = '0'
protocol = QuantumCircuit(2)
protocol.h(0)
protocol.cx(0, 1)
protocol.barrier()

if c == '1':
    protocol.x(0)
if d == '1':
    protocol.z(0)

protocol.cx(0, 1)
protocol.h(0)
protocol.barrier()
protocol.measure_all()
display(protocol.draw(output="mpl"))
```
