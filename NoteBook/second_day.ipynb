# 数学基础
我们需要区分两种不同的数学操作：张量积（Tensor Product）和内积/标量积（Inner Product/Scalar Product）。
1. $|a\rangle$ 与 $|b\rangle$ 的关系：张量积符号： $|a\rangle \otimes |b\rangle$（通常简写为 $|ab\rangle$）含义： 张量积（Tensor Product）。用途： 描述两个独立的量子系统 $X$ 和 $Y$ 组成的复合系统的状态。结果： 复合系统状态空间中的一个态矢量（Ket Vector），其维度是两个子系统维度的乘积。$$|a\rangle \otimes |b\rangle \equiv |ab\rangle \quad \text{（复合系统的态矢）}$$
2. $\langle a|$ 与 $\langle b|$ 的关系：张量积符号： $\langle a| \otimes \langle b|$（通常简写为 $\langle ab|$）含义： 张量积（Tensor Product）。用途： 描述与复合系统状态 $|ab\rangle$ 相对应的对偶空间矢量（Bra Vector）。结果： 复合系统对偶空间中的一个 Bra 矢量，其维度是两个子系统维度的乘积。它是 $|ab\rangle$ 的厄米共轭。$$\langle a| \otimes \langle b| \equiv \langle ab| \quad \text{（复合系统的 Bra 矢）}$$
3. $\langle a|$ 与 $|b\rangle$ 的关系：内积（标量）符号： $\langle a|b\rangle$含义： 内积（Inner Product） 或 标量积（Scalar Product）。用途： 计算两个同一系统（例如都是系统 $X$ 或系统 $Y$）状态矢量之间的重叠程度。结果： 一个标量（复数）。$$\langle a|b\rangle = \text{标量} \quad \text{（如基底正交时，}\langle a|b\rangle = \delta_{ab}\text{）}$$
4. $|a\rangle$ 与 $\langle b|$ 的关系：外积（操作符）符号： $|a\rangle \langle b|$含义： 外积（Outer Product）。用途： 构造作用于同一系统状态空间中的一个线性操作符（Operator）或投影算符（Projection Operator）。结果： 一个操作符/矩阵。$$|a\rangle \langle b| = \text{操作符/矩阵} \quad \text{（将态} |b\rangle \text{映射到} |a\rangle \text{）}$$





### 1. 基础知识：基态顺序

对于两个比特的系统，共有四个基态（Basis States）。量子矩阵总是按照一个固定的顺序排列这些基态，通常是：

$$|00\rangle, \quad |01\rangle, \quad |10\rangle, \quad |11\rangle$$

这个 $4 \times 4$ 矩阵的**四列**对应输入基态的顺序，**四行**对应输出基态的顺序。


#### 1. $|00\rangle$ 状态

它是 $|0\rangle$ 和 $|0\rangle$ 的张量积：
$$|00\rangle = |0\rangle \otimes |0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \otimes \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \begin{pmatrix} 1 \cdot \begin{pmatrix} 1 \\ 0 \end{pmatrix} \\ 0 \cdot \begin{pmatrix} 1 \\ 0 \end{pmatrix} \end{pmatrix} = \begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix}$$

#### 2. $|01\rangle$ 状态

它是 $|0\rangle$ 和 $|1\rangle$ 的张量积：
$$|01\rangle = |0\rangle \otimes |1\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \otimes \begin{pmatrix} 0 \\ 1 \end{pmatrix} = \begin{pmatrix} 1 \cdot \begin{pmatrix} 0 \\ 1 \end{pmatrix} \\ 0 \cdot \begin{pmatrix} 0 \\ 1 \end{pmatrix} \end{pmatrix} = \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix}$$

#### 3. $|10\rangle$ 状态

它是 $|1\rangle$ 和 $|0\rangle$ 的张量积：
$$|10\rangle = |1\rangle \otimes |0\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \otimes \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \begin{pmatrix} 0 \cdot \begin{pmatrix} 1 \\ 0 \end{pmatrix} \\ 1 \cdot \begin{pmatrix} 1 \\ 0 \end{pmatrix} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix}$$

#### 4. $|11\rangle$ 状态

它是 $|1\rangle$ 和 $|1\rangle$ 的张量积：
$$|11\rangle = |1\rangle \otimes |1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \otimes \begin{pmatrix} 0 \\ 1 \end{pmatrix} = \begin{pmatrix} 0 \cdot \begin{pmatrix} 0 \\ 1 \end{pmatrix} \\ 1 \cdot \begin{pmatrix} 0 \\ 1 \end{pmatrix} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}$$



# 2. **纠缠态（Entanglement）**。

通过**反证法（Proof by Contradiction）**，证明了著名的 **贝尔态（Bell State）**之一：

$$\mathbf{|\Phi^{+}\rangle} = \frac{1}{\sqrt{2}}|00\rangle + \frac{1}{\sqrt{2}}|11\rangle$$

**不是一个乘积态（Product State），而是一个纠缠态。**

---

## 这不是一个乘积态

乘积态（Product State）指的是一个多比特状态可以写成各个单比特状态的**张量积**形式。如果 $|\Phi^{+}\rangle$ 是乘积态，那么它应该可以写成：

$$|\Phi^{+}\rangle = |\phi\rangle \otimes |\psi\rangle$$

其中，
* $|\phi\rangle = \alpha |0\rangle + \beta |1\rangle$ 是第一个量子比特的状态。
* $|\psi\rangle = \gamma |0\rangle + \delta |1\rangle$ 是第二个量子比特的状态。

### 步骤一：假设为乘积态并展开

**假设：** 假设 $|\Phi^{+}\rangle$ 是一个乘积态，即：
$$\frac{1}{\sqrt{2}}|00\rangle + \frac{1}{\sqrt{2}}|11\rangle = |\phi\rangle \otimes |\psi\rangle$$

如果我们将右侧的张量积展开，应该得到：
$$|\phi\rangle \otimes |\psi\rangle = (\alpha |0\rangle + \beta |1\rangle) \otimes (\gamma |0\rangle + \delta |1\rangle)$$
$$= \alpha\gamma |00\rangle + \alpha\delta |01\rangle + \beta\gamma |10\rangle + \beta\delta |11\rangle$$

为了使假设成立，展开后的系数必须与左侧的系数相等：
1.  系数 for $|00\rangle$: $\alpha\gamma = \frac{1}{\sqrt{2}}$
2.  系数 for $|01\rangle$: $\alpha\delta = 0$
3.  系数 for $|10\rangle$: $\beta\gamma = 0$
4.  系数 for $|11\rangle$: $\beta\delta = \frac{1}{\sqrt{2}}$

### 步骤二：利用正交性推出矛盾



考虑 $|01\rangle$ 态：
$$\langle 01 | (|\phi\rangle \otimes |\psi\rangle) = 0$$

* 左侧的 $\langle 01 |$ 是一个内积操作。由于 $|01\rangle$ 是一个基态，且在假设的展开式中 $|01\rangle$ 的系数为 $\alpha\delta$。根据**要求 2**，这个系数必须为 $0$。

$$\langle 01 | (|\phi\rangle \otimes |\psi\rangle) = \langle 0 | \phi \rangle \langle 1 | \psi \rangle = (\alpha) \cdot (\delta) = \mathbf{0}$$

这导致：
$$\alpha = 0 \quad \text{或} \quad \delta = 0 \quad (\text{或两者皆是})$$

### 步骤三：导致矛盾 (Contradiction)

现在，我们把 $\alpha=0$ 或 $\delta=0$ 代入**要求 1** ($|00\rangle$ 的系数) 和**要求 4** ($|11\rangle$ 的系数)，图中展示了正是这四个公式导致了矛盾：

#### 如果 $\alpha = 0$:
* 代入要求 1：$\alpha\gamma = 0 \cdot \gamma = 0$。但这与要求 1 $\alpha\gamma = \frac{1}{\sqrt{2}}$ **矛盾**！

#### 如果 $\delta = 0$:
* 代入要求 4：$\beta\delta = \beta \cdot 0 = 0$。但这与要求 4 $\beta\delta = \frac{1}{\sqrt{2}}$ **矛盾**！

---

### 结论：纠缠态的定义

由于对 $|\Phi^{+}\rangle$ 是乘积态的**假设**必然导致**矛盾**，因此：

$$|\Phi^{+}\rangle = \frac{1}{\sqrt{2}}|00\rangle + \frac{1}{\sqrt{2}}|11\rangle$$

**不是一个乘积态，它是一个不可分解的联合状态，即** **纠缠态（Entangled State）**。

**物理意义：** 这两个量子比特 $Q_1$ 和 $Q_2$ 的命运是**锁定**的。如果你测量 $Q_1$ 得到 0，那么 $Q_2$ 必然是 0；如果你测量 $Q_1$ 得到 1，那么 $Q_2$ 必然是 1。它们不能被独立描述。


## 3. **测量公设 (Measurement Postulate)** 和 **波函数坍缩 (Wave Function Collapse)**。

---

### 复合系统的部分测量

提供的图片解释了当我们有一个由两个子系统 **X** 和 **Y** 组成的复合系统 $(X, Y)$ 处于某个量子态 $| \psi \rangle$ 时，如果只对系统 **X** 进行标准基测量，而对 **Y** 不做任何操作，会发生什么。

### 1. 初始量子态的表示

系统 $(X, Y)$ 的整体量子态 $| \psi \rangle$ 处于 **X** 的基态 $\Sigma$ 和 **Y** 的基态 $\Gamma$ 所张成的复合希尔伯特空间中。

$$
| \psi \rangle = \sum_{(a,b) \in \Sigma \times \Gamma} \alpha_{ab} |a b \rangle
$$

* $|a b \rangle = |a \rangle \otimes |b \rangle$ 表示 **X** 处于基态 $|a\rangle \in \Sigma$ 且 **Y** 处于基态 $|b\rangle \in \Gamma$ 的联合状态。
* $\alpha_{ab}$ 是该联合状态的**概率幅 (probability amplitude)**。
* 由于 $| \psi \rangle$ 是一个**归一化 (normalized)** 的量子态，所有概率幅的模方之和必须为 1：$\sum_{a,b} |\alpha_{ab}|^2 = 1$。

### 2. 仅测量系统 X 的概率

当只测量系统 **X** 时，测量结果 **a** 的出现概率 $Pr(\text{outcome is } a)$ 是所有包含该结果 $a$ 的联合状态概率之和。

* 对于一个特定的 $a \in \Sigma$，**X** 处于 $|a\rangle$ 的所有可能状态是 $\{|a b \rangle \mid b \in \Gamma\}$。
* 每个联合状态 $|a b \rangle$ 出现的概率是 $|\langle a b | \psi \rangle |^2 = |\alpha_{ab}|^2$。

因此，测量 **X** 得到结果 $a$ 的总概率是：

$$
Pr(\text{outcome is } a) = \sum_{b \in \Gamma} |\langle a b | \psi \rangle |^2 = \sum_{b \in \Gamma} |\alpha_{ab}|^2
$$

### 3. 初始状态的另一种分解 (施密特分解形式)

为了更清晰地描述测量后的状态变化，可以将初始状态 $| \psi \rangle$ 重新分解，将所有 **Y** 的部分集合起来。

$$
| \psi \rangle = \sum_{a \in \Sigma} |a \rangle \otimes \left( \sum_{b \in \Gamma} \alpha_{ab} |b \rangle \right)
$$

定义一个与 **X** 的测量结果 $a$ 相关的 **Y** 的非归一化状态 $| \phi_a \rangle$:

$$
| \phi_a \rangle = \sum_{b \in \Gamma} \alpha_{ab} |b \rangle
$$

那么初始状态可以写成：

$$
| \psi \rangle = \sum_{a \in \Sigma} |a \rangle \otimes | \phi_a \rangle
$$

### 4. 测量结果的概率 (与范数的关系)

使用 $| \phi_a \rangle$ 的定义，我们可以将测量结果 $a$ 的概率与 $| \phi_a \rangle$ 的**范数 (norm)** 联系起来。

**Y** 的非归一化状态 $| \phi_a \rangle$ 的范数平方是：
$$
\| | \phi_a \rangle \|^2 = \langle \phi_a | \phi_a \rangle = \sum_{b \in \Gamma} |\alpha_{ab}|^2
$$

因此，测量 **X** 得到结果 $a$ 的概率就是 $| \phi_a \rangle$ 的范数平方：

$$
Pr(\text{outcome is } a) = \sum_{b \in \Gamma} |\alpha_{ab}|^2 = \| | \phi_a \rangle \|^2
$$

### 5. 测量后的量子态（波函数坍缩）

**最关键的部分是：** 当对 **X** 进行标准基测量，并得到确定的结果 $a$ 后，根据量子力学的**测量公设 (Measurement Postulate)**，整个复合系统 $(X, Y)$ 的状态会**坍缩 (collapse)** 到一个新的归一化状态。

* **坍缩规则：** 测量 **X** 得到 $a$ 的结果，意味着 **X** 的部分确定地处于状态 $|a\rangle$。
* **后继状态：** 整个系统的后继状态是**与 $a$ 对应的非归一化状态 $|a \rangle \otimes | \phi_a \rangle$ 的归一化形式**。

系统 $(X, Y)$ 坍缩到的新状态 $| \psi' \rangle$ 是：

$$
| \psi' \rangle = \frac{|a \rangle \otimes | \phi_a \rangle}{\| | \phi_a \rangle \|} = |a \rangle \otimes \left( \frac{| \phi_a \rangle}{\| | \phi_a \rangle \|} \right)
$$

这意味着：
1.  **X 的状态：** 确定地变为 $|a\rangle$。
2.  **Y 的状态：** 变为**归一化**后的状态 $\frac{| \phi_a \rangle}{\| | \phi_a \rangle \|}$。

### 总结

对复合系统 $(X, Y)$ **只测量 X** 的操作，其后果是：

1.  **结果**：以概率 $Pr(\text{outcome is } a) = \| | \phi_a \rangle \|^2$ 得到测量结果 $a$。
2.  **状态变化**：整个系统 $(X, Y)$ **瞬时地**从 $| \psi \rangle$ **坍缩**到新的归一化状态 $|a \rangle \otimes \left( \frac{| \phi_a \rangle}{\| | \phi_a \rangle \|} \right)$。

**这个过程是量子纠缠的关键所在：** 即使 **Y** 没有被直接测量，它作为复合系统的一部分，其状态也因为 **X** 的测量结果而**立刻**改变了。如果 $| \psi \rangle$ 是一个**纠缠态 (entangled state)**，**Y** 的后继状态会**强烈依赖**于 **X** 的测量结果 $a$。



## 示例演示：对贝尔态进行部分测量

假设我们的复合系统 $(X, Y)$ 是两个量子比特 $Q_1$ 和 $Q_2$，处于著名的贝尔态：

$$
| \psi \rangle = \frac{1}{\sqrt{2}} |00 \rangle + \frac{1}{\sqrt{2}} |11 \rangle
$$

这里：
* $X = Q_1$（第一个量子比特）
* $Y = Q_2$（第二个量子比特）
* 基态 $\Sigma = \{|0\rangle, |1\rangle\}$, $\Gamma = \{|0\rangle, |1\rangle\}$。
* 概率幅 $\alpha_{00} = 1/\sqrt{2}$, $\alpha_{11} = 1/\sqrt{2}$。
* 所有其他 $\alpha_{ab}$ (即 $\alpha_{01}$ 和 $\alpha_{10}$) 都为 0。

### 第一步：找出 $| \phi_a \rangle$

我们根据 $X$ 的两个可能结果 $a=0$ 和 $a=1$，来计算对应的 $| \phi_a \rangle$ 状态：

#### 1. 当 $a=0$ 时：
$$
| \phi_0 \rangle = \sum_{b \in \{0,1\}} \alpha_{0b} |b \rangle = \alpha_{00} |0 \rangle + \alpha_{01} |1 \rangle
$$
代入概率幅：
$$
| \phi_0 \rangle = \frac{1}{\sqrt{2}} |0 \rangle + 0 \cdot |1 \rangle = \frac{1}{\sqrt{2}} |0 \rangle
$$

#### 2. 当 $a=1$ 时：
$$
| \phi_1 \rangle = \sum_{b \in \{0,1\}} \alpha_{1b} |b \rangle = \alpha_{10} |0 \rangle + \alpha_{11} |1 \rangle
$$
代入概率幅：
$$
| \phi_1 \rangle = 0 \cdot |0 \rangle + \frac{1}{\sqrt{2}} |1 \rangle = \frac{1}{\sqrt{2}} |1 \rangle
$$

原始状态的分解形式是：$| \psi \rangle = |0 \rangle \otimes | \phi_0 \rangle + |1 \rangle \otimes | \phi_1 \rangle$
$$
| \psi \rangle = |0 \rangle \otimes \left( \frac{1}{\sqrt{2}} |0 \rangle \right) + |1 \rangle \otimes \left( \frac{1}{\sqrt{2}} |1 \rangle \right)
$$
这与原始的 $| \psi \rangle$ 表达式一致。

---

### 第二步：计算测量 $X$ 的概率

现在我们计算测量 $X$ 得到 $a=0$ 和 $a=1$ 的概率，利用公式 $Pr(\text{outcome is } a) = \| | \phi_a \rangle \|^2$。

#### 1. 测量结果 $a=0$ 的概率：
$$
Pr(a=0) = \| | \phi_0 \rangle \|^2 = \left\| \frac{1}{\sqrt{2}} |0 \rangle \right\|^2
$$
$$
Pr(a=0) = \left( \frac{1}{\sqrt{2}} \right)^2 \langle 0 | 0 \rangle = \frac{1}{2} \cdot 1 = \frac{1}{2}
$$

#### 2. 测量结果 $a=1$ 的概率：
$$
Pr(a=1) = \| | \phi_1 \rangle \|^2 = \left\| \frac{1}{\sqrt{2}} |1 \rangle \right\|^2
$$
$$
Pr(a=1) = \left( \frac{1}{\sqrt{2}} \right)^2 \langle 1 | 1 \rangle = \frac{1}{2} \cdot 1 = \frac{1}{2}
$$

* **结论：** 测量 $X$ 得到 0 或 1 的概率都是 $1/2$ (50%)。

---

### 第三步：计算测量后的新状态（波函数坍缩）

如果测量 $X$ 得到了结果 $a$，系统 $(X, Y)$ 会坍缩到归一化状态 $\frac{|a \rangle \otimes | \phi_a \rangle}{\| | \phi_a \rangle \|}$。

#### 1. **如果测量结果是 $a=0$：**

系统的后继状态 $| \psi'_{0} \rangle$ 是：
$$
| \psi'_{0} \rangle = \frac{|0 \rangle \otimes | \phi_0 \rangle}{\| | \phi_0 \rangle \|} = \frac{|0 \rangle \otimes \left( \frac{1}{\sqrt{2}} |0 \rangle \right)}{\sqrt{1/2}}
$$
由于 $\sqrt{1/2} = 1/\sqrt{2}$，它们相消：
$$
| \psi'_{0} \rangle = |0 \rangle \otimes |0 \rangle = |00 \rangle
$$

#### 2. **如果测量结果是 $a=1$：**

系统的后继状态 $| \psi'_{1} \rangle$ 是：
$$
| \psi'_{1} \rangle = \frac{|1 \rangle \otimes | \phi_1 \rangle}{\| | \phi_1 \rangle \|} = \frac{|1 \rangle \otimes \left( \frac{1}{\sqrt{2}} |1 \rangle \right)}{\sqrt{1/2}}
$$
同理，相消后得到：
$$
| \psi'_{1} \rangle = |1 \rangle \otimes |1 \rangle = |11 \rangle
$$



这个例子完美地展示了**量子纠缠**和**波函数坍缩**：

1.  **强关联性**：在测量之前，$X$ 和 $Y$ 都没有确定的状态，它们是互相纠缠的。
2.  **瞬时坍缩**：一旦我们测量了 $X$ 得到 0，虽然我们没有对 $Y$ 做任何操作，但 $Y$ 的状态**立即**、**确定地**坍缩到了 $|0\rangle$。同样地，如果 $X$ 测到 1， $Y$ 也立即坍缩到 $|1\rangle$。


创建两个量子比特形成的系统，$$|0\rangle \otimes |1\rangle$$

```python
from qiskit.quantum_info import Statevector, Operator
from numpy import sqrt

zero = Statevector.from_label('0')
one = Statevector.from_label('1')
psi = zero.tensor(one)
display(psi.draw('latex'))
```

接下来以两个带相位和旋转的两个qbit为例 $$ |+\rangle  和|-i\rangle $$

```python
plus = Statevector.from_label('+')
minus_i = Statevector.from_label('l')
phi = plus.tensor(minus_i)
display(phi.draw('latex'))
#or use '^' operator replacing 'tensor' method
display((plus ^ minus_i).draw("latex"))
```

算子也可以使用这个类的tensor

```python
H = Operator.from_label('H')
X = Operator.from_label('X')
HX = H.tensor(X)
display(HX.draw('latex'))
# or
display((H  ^ X).draw("latex"))
```

接下来对$|\phi\rangle$ 附加算子 



```python
display(phi.evolve(HX).draw("latex"))
```
