# Quantum information Limitations
Before we introduce the advantage of quantum information, we should know what it can't do. So this section mainly introduces the Irrelevance of Global Phase ,  No-Cloning Theorem and imperfect Discriminability of Non-Orthogonal States.

## 1.Irrelevance of Global Phase.
Explanation of global phase: Imagine we have two basis vectors $|\phi \rangle$ and $|\psi \rangle$. We define $\alpha = e^{i\theta}$, and let $|\phi \rangle = \alpha |\psi \rangle$(while $\theta$ is a real number).  The vector $|\phi \rangle$ and $|\psi \rangle$ differ by a global phase $\theta$.

* Measurement: Since $|\alpha|^2 = 1$,the probabilty of any measuring outcome for $|\phi \rangle$  is equal to measuring that of $|\psi \rangle$.
* Unitary Evolution: If we apply a unitary operator to state $|\phi \rangle$:
$$
U|\phi \rangle  = U(\alpha |\psi \rangle) =  \alpha U|\psi \rangle
$$
The resulting state still differs from $U|\phi \rangle$ only by the phase factor $\alpha$.
We can conclude that the global phase $\theta$ doesn't impact the measuring result.

## 2. No-cloning theorem
The No-cloning theorem states that we can't make a identical copy of an arbitrary unknown quantum state.

### Proof (by contradiction)
Supposed we have a source qubit in state $|\phi \rangle $ and a target qubit in blank state $|\psi \rangle$, we want to design a unitory oparator U to copy $|\phi \rangle $ onto $|\psi \rangle$:
$$
|\phi \rangle \otimes |\psi \rangle  -> |\phi \rangle \otimes |\phi \rangle 
$$
Assume the clone machine works for two orthogonal basis states,$|a \rangle $ and  $|b \rangle $:
\begin{cases}
U(|a \rangle \otimes |\psi \rangle)  = |a \rangle \otimes |a \rangle \\
U(|b \rangle \otimes |\psi \rangle)  = |a \rangle |b \rangle
\end{cases}
Now, consider a superposition state $|\psi \rangle = |a \rangle + |b \rangle$. A contradiction arise when we compare the results derived from linearity  versus the definition of cloning:

\begin{align}
U(|a \rangle \otimes |\psi \rangle + |b \rangle \otimes |\psi \rangle) = |a \rangle \otimes |a \rangle + |b \rangle \otimes |b \rangle \\

\begin{aligned}
U((|a \rangle + |b \rangle) \otimes |\psi \rangle) &= (|a \rangle + |b \rangle) \otimes (|a \rangle + |b \rangle)
\end{aligned}\\

\end{align}

Comparing Equation (1) and (2), we see that they are not equal .
This implies that a general unitary cloning operation violates linearity, which is a fundamental property of quantum mechanics. Thus, general quantum cloning is impossible.


## Non-orthogonal states cannot be perfectly discriminated
Two quantum states $|\phi \rangle$ and $|\psi \rangle$, are non-orthogonal($\langle \phi|\psi \rangle \neq 0$), which means they can't be distinguished perfectly.

### Proof by Contradiction(necessity):
Suppose there exists a quantum circuit that can distinguish these two states, we assume that
* If the input is $|\phi \rangle$,the measurement outcome is always '1'.
* If the input is $|\psi \rangle$,the measurement outcome is always '0'.

The evolved composite system can be represented as:

$$U(|0\cdots0\rangle|\psi\rangle) = |\gamma_0\rangle|0\rangle + |\gamma_1\rangle|1\rangle$$
If the circuit perfectly identifies $|\psi \rangle$ (always outputs '0'), then the amplitude of outcome '1' must be zero($\gamma_1 = 0$).Thus:
$$
U(|0\cdots0\rangle|\psi\rangle) = |\gamma_0\rangle|0\rangle.
$$
Inverting the unitary U, we get :
$$
|0\cdots0\rangle|\psi\rangle = U^\dagger(|\gamma_0\rangle|0\rangle). \ \ \   (1)$$ 
Similarly, for the state $|\phi \rangle$, the outcome is always '1', so:
$$
|0\cdots0\rangle|\phi\rangle = U^\dagger(|\delta_1\rangle|1\rangle). \ \ \ (2)$$
Now, let's take the inner product of eqaution (1) and (2):

Right-hand side:
$$(\langle\gamma_0|\langle 0|) U U^\dagger (|\delta_1\rangle|1\rangle) = (\langle\gamma_0|\langle 0|)(|\delta_1\rangle|1\rangle) = \langle\gamma_0|\delta_1\rangle\langle 0|1\rangle = 0.$$
Left-hand side:
$$
0 = (\langle 0\cdots0|\langle \psi|)^\dagger (|0\cdots0\rangle|\phi\rangle) = \langle 0\cdots0|0\cdots0\rangle \langle \psi|\phi \rangle = \langle \psi|\phi \rangle.
$$
Contradiction:
This implies that $langle\gamma_0|\delta_1\rangle$. However this contradicts our initial assumption that $|\phi \rangle$ and $|\psi \rangle$ are non-orthogonal.Therefore, such a perfect discrimination circuit cannot exist.

### proof by sufficiency
Next, we prove the converse: if two states are orthogonal, they can be perfectly discriminated.
We define a pair orthogonal vectors  $|\phi \rangle$ and $|\psi \rangle$.And define a projective measurement for the proof:
 $$ \{ P_1 = |\phi\rangle\langle\phi|, \quad P_2 = \mathbb{I} - |\phi\rangle\langle\phi| \} $$
Case 1: If the state is $|\phi \rangle$:

The probability of obtaining the first outcome $(P_1)$ is :
$$ \left\| |\phi\rangle \langle\phi| |\phi\rangle \right\|^2 = \left\| |\phi\rangle \langle\phi|\phi\rangle \right\|^2 = \left\| |\phi\rangle \right\|^2 = 1$$
The probability of obtaining the second outcome $(P_2)$ is 0.

Case 2:  If the state is $|\psi \rangle$:

The probability of obtain the first outcome $(P_1)$ is :
$$ \left\| |\phi\rangle \langle\phi| |\psi\rangle \right\|^2 = \left\| |\phi\rangle \langle\phi|\psi\rangle \right\|^2 = \left\| \mathbf{0} \right\|^2 = 0$$
Therefore, the probability of obtaining the second outcome  $(P_2)$ is 1.

Since the outcomes are deterministic and distinct, the states can be perfectly discriminated.
