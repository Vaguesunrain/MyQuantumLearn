# Linear algebra basics

## Two qubits represent

A single qubit can be mathematically described as a superpositon state $\alpha |0\rangle + \beta |1\rangle$,where $\alpha |0\rangle \  \beta |1\rangle$ form the basis. A separable state of two qubits is described as the tensor product of two individual qubits.  
$$ ( a|0\rangle +  b|1\rangle) \otimes ( c|0\rangle + d |1\rangle) =  ac|00\rangle + ad|01\rangle + bc|10\rangle +  bd|11\rangle $$
The values of '$a,b,c,d$' are complex coefficients satisfying the normalization  condition $\left(|a|^2 + |b|^2\right) \cdot \left(|c|^2 + |d|^2\right) = 1$.These two qubits span a 4-dimensional Hilbert space.
And how about the entangled state? 

### Entangled state
We can demonstrate the Bell state is not a product state but an entangled state using proof by contradiction.
$$\mathbf{|\Phi^{+}\rangle} = \frac{1}{\sqrt{2}}|00\rangle + \frac{1}{\sqrt{2}}|11\rangle$$

Supposed that's a product state. Then the coefficients must satisfy:
1.  for $|00\rangle$: $ac = \frac{1}{\sqrt{2}}$
2.  for $|01\rangle$: $ad = 0$
3.  for $|10\rangle$: $bc = 0$
4.  for $|11\rangle$: $bd = \frac{1}{\sqrt{2}}$

The condition above is  conflicting, So the entangled state shares a strong bond between two qubits.Measuring qubit A instantly determines the outcome of measuring qubit B(They can't be described independently).

## Operators
Just imagine if we want to apply an operator to a qubit, what must we  keep in mind to maintain the mathematical consistency? 
Considering a single qubit is described by a two dimensional vector, if we operate on this qubit(apply a gate to it), the operator must be a 2* 2 square matrix.Similarly,for a two-qubit system, the oparetor must be a 4*4 square matrix. 
(However,this is not enough,the operator must also be a Unitary matrix. I will explain it later on.)

### Matrix classification


| Name | Definition | Significance |
| :---: | :---: | :--- |
| Hermitian | $M = M^\dagger$ | **Observable**|
| Unitary | $U^\dagger U = \mathbb{I}$ | **Quantum Gate**|
| Projective | $P = P^\dagger$  &  $P^2 = P$ | **Measurement**|
| Stochastic | $M_{ij} \ge 0$  &   $\sum_{j} M_{ij} = 1$ | **classical probability transition**|
| Density Matrix | $\rho = \rho^\dagger$  &  $\text{Tr}(\rho) = 1$  &  $\rho \ge 0$  | **quantum state**|
| Trace-Preserving | $\sum_{i} K_i^\dagger K_i = \mathbb{I}$ | **quantum channels/processes**|
| Completely Positive and Trace-Non-Increasing | $\sum_{i} K_i^\dagger K_i \le \mathbb{I}$ | **non-unitary evolution**|
| Involution/Self-Inverse | $M^2 = \mathbb{I}$ | **periodic operations**|
| Permutation | $P_{ij} \in \{0, 1\}$  &  Every row or column has just one value of 1 | **reversible classical state swaps**|

If a system is in one of the Hermitain matrix $M$'s eigenvector $|v\rangle$,measuring $M$ yields the outcome corresponding to the eigenvalue $\lambda$.


* 2. Unitary 
Unitary operators make the vector rotate on the Bloch Sphere(what's the Bloch Sphere? You can simplify this as a normal Sphere,and a vector starts from the center ends at the surface.),but maintain the vector's norm(length) unchanging.Since the length is preserved, the total probability remains conserved.It's extremely important for a quantum state.Furthermore ,it's reversible meaning you can use $U^\dagger$ to undo the operation.(The gates introduced later--X,Z,H,CNOT-- they are all unitary operators.)
  


* 3. Projective 
magine a beam of light shining on an object, casting a shadow on the ground.
 This intuitively explains $P^2 = P$: If the shadow has **already been projected** onto the ground, no matter how many times you project it again, the shadow never changes.
It is the same when you **measure a quantum system**. Once you get a result (e.g., 0), the result never changes **even if you measure it again**. (I will explain this systematically later.)


* 4. Stochastic 
**Think of tossing a coin.** It has a **50% chance of landing on Heads and 50% on Tails**.
A Stochastic Matrix describes how **probabilities change** as time goes by.
 **In contrast to** Unitary operators, Stochastic matrices **deal with probabilities** (real numbers, which can only be added), **whereas** Unitary operators deal with **probability amplitudes** (complex numbers, which can interfere and cancel out).
 In quantum computing, when the SNR is very low (or **decoherence** occurs), quantum behavior **degenerates**. The system then evolves like a Stochastic matrix, **losing** the quantum advantage.

## Bloch Sphere

### Axis
* Z axis(classical axis):
    * The north pole($+Z$):$∣0⟩$,the calssical bit 0.
    * The south pole($-Z$):$∣1⟩$,the calssical bit 1.
    * The projection of the Qubit onto the Z-axis determines the measurement probability. The closer the state is to the North Pole, the greater the probability of measuring 0.
 
* X axis(axis of superposition)
    * $+X$ :$|+\rangle = \frac{|0\rangle+|1\rangle}{\sqrt{2}}$.
    * $-X$ :$|-\rangle = \frac{|0\rangle-|1\rangle}{\sqrt{2}}$.
    * Represents a perfect superposition created by the Hadamard gate.

* Y axis(axis of imaginary  Phase)
    * $+Y$:$|i\rangle = \frac{|0\rangle+i|1\rangle}{\sqrt{2}}$.($|R\rangle$)
    *  $-Y$:$|-i\rangle = \frac{|0\rangle-i|1\rangle}{\sqrt{2}}$.($|L\rangle$)
    * Represents a state with a purely imaginary relative phase($e^{i\pi/2} = i$).
    

```python

import numpy as np
import matplotlib.pyplot as plt

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram

# %matplotlib inline 

print("env is ready")
```

```python



psi_super = Statevector.from_label('0').evolve(np.array([[1, 1], [1, -1]]) / np.sqrt(2))

fig1 = plot_bloch_multivector(psi_super, title="State |+> (Equator)")
print("vector in equator (Z=0)" )
display(fig1) 
plt.close(fig1)  

qc = QuantumCircuit(1)
qc.h(0)          
qc.measure_all() 

simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit, shots=1024).result()
counts = result.get_counts()

print("\n=== statistic (1024 shots) ===")
print(f"Measurement results: {counts}")
print("Note: You can see that the counts for 0 and 1 are very close, each about 50%.")


fig2 = plot_histogram(counts, title="Measurement Statistics (H Gate)")
display(fig2)
plt.close(fig2)  
```

```python

# We let vector rotate pi/3 (60 degrees) around Y axis
theta = np.pi / 3 

ry_matrix = np.array([[np.cos(theta/2), -np.sin(theta/2)],
                      [np.sin(theta/2),  np.cos(theta/2)]])
psi_tilted = Statevector.from_label('0').evolve(ry_matrix)
fig1 = plot_bloch_multivector(psi_tilted, title=f"Rotated by {np.degrees(theta):.1f} deg")
display(fig1)
plt.close(fig1)

qc2 = QuantumCircuit(1)
qc2.ry(theta, 0) 
qc2.measure_all()

compiled_circuit2 = transpile(qc2, simulator)
result2 = simulator.run(compiled_circuit2, shots=1024).result()
counts2 = result2.get_counts()

print("\n=== statistic ===")
print(f"Measurement results: {counts2}")
print("Theoretical prediction: 0 should be ~75%, 1 should be ~25%")


fig2 = plot_histogram(counts2, title="Statistics (Bias towards 0)")
display(fig2)
plt.close(fig2)
```

## Pauli gates


### I. Single-Qubit Gates

| name | notation | matrix form ($M$) | significance | usage |
| :---: | :---: | :---: | :--- | :--- |
| **Pauli-X gate** | $X$ | $\begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$ |  Bit-Flip.Rotate  $180^\circ$ around $X$. $\|0\rangle \leftrightarrow \|1\rangle$. |Like a classical NOT gate. |
| **Pauli-Y gate** | $Y$ | $\begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}$ | Rotate  $180^\circ$ around $Y$.Combined bit and phase flip. | quantum state control. |
| **Pauli-Z gate** | $Z$ | $\begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$ | Phase-Flip.Rotate  $180^\circ$ around $Z$. $\|1\rangle \to -\|1\rangle$。 | Correct,phase encodeing. |
| **Hadamard gate** | $H$ | $\frac{1}{\sqrt{2}}\begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}$ | Create Superposition, Mapping $\|0\rangle, \|1\rangle$ to $\|+\rangle, \|-\rangle$。 | Make entanglement,basis transformation ($X$ axis $\leftrightarrow Z$ axis). |
| **S gate** | $S$ | $\begin{pmatrix} 1 & 0 \\ 0 & i \end{pmatrix}$ | $\pi/2$ phase gate.Rotate  $90^\circ$ around $Z$. ($S^2 = Z$) | phase alignment ，Clifford group operations.|
| **T gate** | $T$ | $\begin{pmatrix} 1 & 0 \\ 0 & e^{i\pi/4} \end{pmatrix}$ | $\pi/4$ phase gate。Rotate  $45^\circ$ around $Z$. (Not Clifford gate) |  Essential for universal quantum computation. |

| **Toffoli gate** | $CCNOT$ |  (8x8) | When both control-bits are  $\|1\rangle$ ，apply not to the target $X$  | **Universal Reversible Computing**; Implementing classical logic (AND). |
