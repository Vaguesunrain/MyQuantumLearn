# Measurement Postulate and Wave Function Collapse
It's a very common scenario that we perform a measurement on just one subsystem within a composite system.For simplicity, we focus on a composite system consisting of subsystems X and Y.  Let's see what happens if we measure system X but leave Y unmeasured.

## 1.Initial state 

The overall quantum state $| \psi \rangle$  of system  $(X, Y)$ resides in the composite Hilbert space spanned by the basis states  $\Sigma$ of $X$ and $\Gamma$ of $Y$.
$$
| \psi \rangle = \sum_{(a,b) \in \Sigma \times \Gamma} \alpha_{ab} |a b \rangle
$$
* Here $\alpha_{ab}$ is the probability amplitude associate with basis state $|ab\rangle$, which satisfies the normalization condition  $\sum_{a,b} |\alpha_{ab}|^2 = 1$.

##  2.Probability of measuring $X$

We define $Pr(\text{outcome is } a)$ as the total probability that measuring $X$ yields $a$.

For a specific outcome  $a \in \Sigma$, the measurement corresponds to projection onto the subspace spanned by set $\{|a b \rangle \mid b \in \Gamma\}$.The probability of finding the system in the specific joint basis state $|a b \rangle$  is $|\langle a b | \psi \rangle |^2 = |\alpha_{ab}|^2$.

For clarity, let's consider the case where $a=0$.

$$
Pr(\text{outcome is } 0) = \sum_{b \in \Gamma} |\langle 0 b | \psi \rangle |^2 = \sum_{b \in \Gamma} |\alpha_{0b}|^2
$$

## Note: explain
It's very complex, isn't it?
Take an example: Suppose system $X$ is a specific state vector, resides in a space spanned by several orthogonal basis vectors.It's easy to understand the probability of measuring 0 by looking at its coefficient(or the projection on the Z axis of Bloch Sphere). 
Now, associate $X$ with $Y$($Y$ also consist of several orthogonal vectors). We can represent the new system by $X \otimes Y$(a higher-dimensional Hilbert space). The probability of observing outcome 0 is dispersed into all the subspaces(e.g.,$|0 b_1 \rangle$,$|0 b_2 \rangle$,...). 
The operation $|\langle a b | \psi \rangle |$ essentially use the bra $\langle ab|$ to extract the specific coefficient(amplitute) of the basis state $|a\rangle$ or $|b\rangle$ from the composite state $| \psi \rangle$. 



## 3.Rewriting Initial State (Preparation for Partial Measurement)
$$
| \psi \rangle = \sum_{a \in \Sigma} |a \rangle \otimes \left( \sum_{b \in \Gamma} \alpha_{ab} |b \rangle \right)  \tag{1}
$$
We define a unnormalized state  $| \phi_a \rangle$ associated with  measuring outcome  a:
$$
| \phi_a \rangle = \sum_{b \in \Gamma} \alpha_{ab} |b \rangle  \tag{2}
$$
According to (1),(2),the total state can be rewritten as:
$$
| \psi \rangle = \sum_{a \in \Sigma} |a \rangle \otimes  |\phi_a \rangle
$$

And we can associate result a with the norm of the vector $|\phi_a \rangle$:s
$$
\| | \phi_a \rangle \|^2 = \langle \phi_a | \phi_a \rangle = \sum_{b \in \Gamma} |\alpha_{ab}|^2
$$
Thus, the probability of obtaining outcome a  is equal to :
$$
Pr(\text{outcome is } a) = \sum_{b \in \Gamma} |\alpha_{ab}|^2 = \| | \phi_a \rangle \|^2
$$

## 4.Quantum state after measuring(Wave Function collapse)
According to Measurement Postulate, when we perform basis measurement on $X$, and obtain a confirmed outcome, The composite system will collapse into a new normalized state.

If we obtain the outcome of a, it meanings that system $X$ is exactly in the state $|a\rangle$.Correspondingly, the total system  evolves to a normalized form  of the vector $|a \rangle \otimes | \phi_a \rangle$ .The new state is represented as :
$$
| \psi' \rangle = \frac{|a \rangle \otimes | \phi_a \rangle}{\| | \phi_a \rangle \|} = |a \rangle \otimes \left( \frac{| \phi_a \rangle}{\| | \phi_a \rangle \|} \right)
$$
Now , $Y$ is a normalized state $\frac{| \phi_a \rangle}{\| | \phi_a \rangle \|}$.
